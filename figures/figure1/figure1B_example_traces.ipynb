```python
import numpy as np
import matplotlib.pyplot as plt
from sys import path
path.insert(1, r"C:\Users\caele\Desktop\Spring2025\miller-lab-final")
from model.model import simulateISP
```

```python
thetaE = 2
thetaI = 5.5
WEE =  2.78415173362416
WEI =  4.826424331460146
WIE =  -0.6920758668123493
WII =  -0.8632121657324914

dt = 0.1e-3
```

```python
plt.style.use('seaborn-v0_8-paper')
```

```python
duration = 12
rE, rI = simulateISP(dt, duration, 100, 10e-3, 10e-3, WEE, WEI, WIE, WII, thetaE, thetaI, np.zeros(int(duration/dt)), np.zeros(int(duration/dt)),
                     rE0=5, rI0=10)
plt.plot(np.arange(0, duration, dt), rE, label='Excitatory', c= "#7879ff")
plt.plot(np.arange(0, duration, dt), rI, label='Inhibitory', c='#F58E89')
plt.legend()
plt.xlabel('Time (s)')
```
