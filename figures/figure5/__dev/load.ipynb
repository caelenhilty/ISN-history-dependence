```python
r = np.load('reliabilities_1.npy')
r
```

```python
import matplotlib.pyplot as plt
amp_range = np.logspace(0, 2, 5)
dur_range = np.linspace(0, 0.1, 5)
dur_mesh, amp_mesh = np.meshgrid(dur_range, amp_range) # amp on y-axis, duration on x-axis
dur_flat, amp_flat = dur_mesh.ravel(), amp_mesh.ravel()
fig, ax = plt.subplots(layout='constrained')
c = ax.pcolormesh(dur_mesh, amp_mesh, r.reshape(dur_mesh.shape), cmap='viridis', shading='auto')
ax.set_yscale('log')
ax.set_xlabel(r"$\tau_{dur}$")
ax.set_ylabel(r"$I_{app}$")
x_offset = (max(dur_range) - min(dur_range))/(2*len(dur_range))
ax.set_xlim(min(dur_range) - x_offset, max(dur_range) + x_offset)
y_offset = (np.log10(max(amp_range)) - np.log10(min(amp_range))) / (2 * len(amp_range))
ax.set_ylim(10**(np.log10(min(amp_range)) - y_offset), max(amp_range))
cbar = fig.colorbar(c, ax=ax)
cbar.set_label('Accuracy')

plt.show()
```
